```python
import pandas as pd
import MeCab
import json
```

```python
repl_df = pd.read_csv("Production_train_sentence_with_repl_compound.csv")
repl_df.head()
```

```python
m_parser = MeCab.Tagger("-u /Users/dontsentouin/mecab-user-dic/compound.dic -Owakati")
```

```python
def parse(s):
    words = []
    node = m_parser.parseToNode(s)
    while node:
        if len(node.surface) is 0:
            node = node.next
            continue

        words.append(node.surface)
        node = node.next
    
    return words
```

```python
repl_df = repl_df.assign(words = repl_df.sentence.apply(lambda x: parse(x)))
repl_df.head()
```
